### 4. Clustering

Before conducting clustering, we will create a new dataframe and drop Position, Goal Difference, and Points because these variables might not be too meaningful in the cluster process. 
- Goal Difference is a mathematical solution between the subtraction of goals conceded by goals scored. 
- Position, and Points have a role similar to an index.

```python
#create a new dataframe by copying the list
filtered_points_table = points_table.loc[:, ["Team", "W", "D","L", "GF", "GA"]].copy()
filtered_points_table = filtered_points_table.reset_index(drop = True)
filtered_points_table.head(5)
```

```python
#select only numerical variables (i.e drop column "Team")
numerical_df = filtered_points_table.select_dtypes(include='number')
numerical_df.head(5)
```

#### 4.1 Choose the best K 

The below code is to create a plot to visualize the "elbow point" which easily help choose the optimal number of clusters in order to strike a balance between having enough clusters to capture data patterns and avoiding too many clusters that overfit the data.

```python
#calculate average_dispersion for different values of k
average_dispersion = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(numerical_df)
    wcss = kmeans.inertia_
    average_dispersion.append(wcss / len(filtered_points_table))  # Calculate average dispersion

#create the elbow plot with average dispersion on the y-axis
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), average_dispersion, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Average Dispersion')
plt.title('Elbow Method for Optimal k (Average Dispersion)')
plt.show()
```

**Comment:**

- The plot suggests that we should choose K = 3

#### 4.2 Standardize The Numerical Data

Standardization helps in improving the clustering results, especially when numerical features have different scales.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_df_scaled = scaler.fit_transform(numerical_df)
```

#### 4.3 Apply kMeans clustering

```python
import os

os.environ['OMP_NUM_THREADS'] = '1'

k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
cluster_labels = kmeans.fit_predict(numerical_df_scaled)
```

```python
#Add the cluster labels to the original DataFrame:
filtered_points_table['Cluster'] = cluster_labels
cluster_means = filtered_points_table.groupby('Cluster').mean(numeric_only=True)
print(cluster_means)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
numerical_df_2d = pca.fit_transform(numerical_df_scaled)

plt.scatter(numerical_df_2d[:, 0], numerical_df_2d[:, 1], c=cluster_labels, cmap='rainbow')
plt.title('kMeans Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
x_column = 'GF'
y_column = 'GA'
plt.scatter(filtered_points_table[filtered_points_table['Cluster'] == 0][x_column], 
            filtered_points_table[filtered_points_table['Cluster'] == 0][y_column], label='Cluster 0', c='blue')

plt.scatter(filtered_points_table[filtered_points_table['Cluster'] == 1][x_column], 
            filtered_points_table[filtered_points_table['Cluster'] == 1][y_column], label='Cluster 1', c='red')

plt.scatter(filtered_points_table[filtered_points_table['Cluster'] == 2][x_column], 
            filtered_points_table[filtered_points_table['Cluster'] == 2][y_column], label='Cluster 2', c='green')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()
plt.title('Cluster Visualization')
plt.show()
```

In upcoming task, we are going to recreate the dataframe, to display who Manchester United play against with, and whether they play home or away game.

Note that it is necessary to switch column "Result". To be clear,
- Originally, Manchester United win 2:1 over West Ham (displayed as West Ham 1:2 Manchester United, index 4, row 5).
- When we take this input into a dataframe without switching, it is still displayed as 1:2, which means Manchester United lose.
- Therefore, we solve this issue by switching all away matches to receive the correct result.

For better understanding, we will create an incorrect dataframe. 

```python
manutd_df = pd.DataFrame(columns=['Date', 'Match', 'Home/Away', 'Result'])

# Iterate through the rows of the original DataFrame
for index, row in manu_match.iterrows():
    if row['HomeTeam'] == 'Manchester United':
        # This row represents a home match for Manchester United
        match = f" {row['AwayTeam']}"
        manutd_df = manutd_df.append({'Date': row['Date'], 'Match': match, 'Home/Away': 'Home', 'Result': row['Result']}, ignore_index=True)
    elif row['AwayTeam'] == 'Manchester United':
        # This row represents an away match for Manchester United
        match = f"{row['HomeTeam']}"
        manutd_df = manutd_df.append({'Date': row['Date'], 'Match': match, 'Home/Away': 'Away', 'Result': row['Result']}, ignore_index=True)

manutd_df[(manutd_df["Match"] == "West Ham United") & (manutd_df["Home/Away"] == "Away")]
```

It should be 2:1 as mentioned before, instead of 1:2.
Now we create a correct dataframe

```python
manutd_df = pd.DataFrame(columns=['Date', 'Match', 'Home/Away', 'Result'])

# Iterate through the rows of the original DataFrame
for index, row in manu_match.iterrows():
    if row['HomeTeam'] == 'Manchester United':
        # This row represents a home match for Manchester United
        match = f" {row['AwayTeam']}"
        manutd_df = manutd_df.append({'Date': row['Date'], 'Match': match, 'Home/Away': 'Home', 'Result': row['Result']}, ignore_index=True)
    elif row['AwayTeam'] == 'Manchester United':
        # This row represents an away match for Manchester United
        match = f"{row['HomeTeam']}"
        
        # Switch the numbers in the 'Result' column for away matches
        result_parts = row['Result'].split(':')
        reversed_result = f"{result_parts[1]}:{result_parts[0]}"
        
        manutd_df = manutd_df.append({'Date': row['Date'], 'Match': match, 'Home/Away': 'Away', 'Result': reversed_result}, ignore_index=True)

manutd_df.head(10)
```

```python
#split column Result into Scored and Conceded columns, delimited by colon ":"
manutd_df[['Scored', 'Conceded']] = manutd_df['Result'].str.split(':', expand=True)
manutd_df.drop("Result", axis=1, inplace=True)
manutd_df.head(10)
```

```python
manutd_df.dtypes #check data types
```

```python
#convert Scored and Conceded into int values
manutd_df["Scored"] = manutd_df["Scored"].astype("int64")
manutd_df["Conceded"] = manutd_df["Conceded"].astype("int64")
manutd_df.dtypes
```

Next, based on the scored and conceded, we create a new column result to indicate whether the result of a match is a win, draw, or lose
- If Scored > Conceded: Manchester United wins
- If Scored = Conceded: Manchester United draws
- If Scored < Conceded: Manchester United loses

```python
results = [] #initialize an empty list

#iterate through the DataFrame rows
for index, row in manutd_df.iterrows():
    if row['Scored'] > row['Conceded']:
        result = 'W'
    elif row['Scored'] < row['Conceded']:
        result = 'L'
    else:
        result = 'D'
    results.append(result)

# Add the "Result" column to the DataFrame
manutd_df['Result'] = results
manutd_df.head(10)
```

```python
result_counts = manutd_df["Result"].value_counts()
sns.barplot(x = result_counts.index, y = result_counts, data = manutd_df,color = grey) 

plt.title("Manchester United Match Results in EPL 2021-2022")
plt.xlabel("Match Result")
plt.ylabel("Count")

plt.show()
```

**Comment**

- Another issue of Manchester United in season 2021-2022 is that they have quite a lot of lose matches (around 12/38 matches).
- Their number of defeats are even higher than their draw games.

In the coming section, we are going to draw some graphs to see the performance of Manchester United given from the start until the end of the season

```python
manutd_df["Goal Difference"] = manutd_df["Scored"] - manutd_df["Conceded"] #create a new column Goal Difference
manutd_df.head(5)
```

```python
manutd_df['Date'] = pd.to_datetime(manutd_df['Date']) #covert to YYYY-MM-DD
manutd_df.head(10)
```

```python
import matplotlib.dates as mdates
plt.figure(figsize=(10, 6))

compare_date = pd.to_datetime('2022-01-05')

# Use boolean indexing to find the first match after the specified date
first_match_after_date = manutd_df[manutd_df['Date'] > compare_date].iloc[0]['Date']
condition_date = pd.to_datetime(first_match_after_date)

home_goals = manutd_df[manutd_df['Home/Away'] == 'Home']
away_goals = manutd_df[manutd_df['Home/Away'] == 'Away']
plt.plot(home_goals['Date'], home_goals['Goal Difference'], label='Home',color = "#1E90FF")
plt.plot(away_goals['Date'], away_goals['Goal Difference'], label='Away',color = grey)

# Add a vertical line at the condition date
plt.axvline(x=condition_date, color=red, linestyle='--', label='Second Leg Start')

# Format the date labels to display days, months, and years
date_format = mdates.DateFormatter('%Y-%m-%d')
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()  # Rotate x-axis labels for better readability

plt.xlabel('Date')
plt.ylabel('Goal Difference')
plt.title('Goal Difference Throughout The Season')
plt.legend()

plt.xlabel('Date')
plt.ylabel('Goal Difference')
plt.title('Football Season Goal Difference')
plt.legend()
plt.grid(False)

plt.show()
```

**Comment:**

- Overall, Manchester United have great start at the beginning of the season.
- In the first leg of the season, they have few loses (looking at points below 0)
- In the second leg of the season, their performance decreases. Most of away games in the second half of the season are defeats.
